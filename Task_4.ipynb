```python
# Install required library
!pip install transformers torch -q

from transformers import pipeline

# Load text generation model
generator = pipeline("text-generation", model="gpt2")

# ------------------------------
# Input: Intern Profile & Role
# ------------------------------

intern_profile = """
Name: Kashan
Skills: Python, Machine Learning, Data Science, NLP
Experience: Built ML and NLP projects
"""

job_description = """
Role: AI Intern
Responsibilities: Work on healthcare AI models, build predictive systems,
assist in NLP-based analytics.
"""

# ------------------------------
# Prompt Engineering
# ------------------------------

prompt = f"""
You are an expert technical interviewer.

Based on the following intern profile and job description,
generate:

1. 5 Technical Interview Questions
2. 5 Behavioral Interview Questions

Intern Profile:
{intern_profile}

Job Description:
{job_description}

Questions:
"""

# ------------------------------
# Generate Questions
# ------------------------------

output = generator(
    prompt,
    max_length=400,
    num_return_sequences=1,
    temperature=0.7
)

print(output[0]['generated_text'])
```
